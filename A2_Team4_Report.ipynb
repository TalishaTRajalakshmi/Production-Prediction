# 3. Assumptions


<hr style="height: 4px; background-color: #2B547E;">
</h1>

In developing our forecasting model for oil and gas production in New Mexico, we based our analysis on key assumptions to streamline focus and enhance model efficiency:

- **Geological Formations:** Our analysis specifically focused on the geological formations we identified as significant for predicting production trends. This decision was based on the premise that these selected formations have a pronounced impact on production capacities, allowing for a more targeted and efficient modeling process.

- **Exclusion of Boring Depths:** The model does not account for vertical and horizontal boring depths. This assumption was made under the rationale that, while depth can influence production, the available data and the model's scope aimed to highlight the influence of geological characteristics over mechanical variables.

- **Latitude and Longitude Omission:** Geographic coordinates (latitude and longitude) were not included in our analysis. This decision was driven by the aim to generalize the model's applicability across New Mexico without narrowing the focus to specific locations, thereby emphasizing broader geological impacts over spatial variability.

These assumptions allowed for a targeted approach, optimizing the model's relevance and applicability within the constraints of available data and project scope.

### Cleaning functions

```python
# function to find values that are not floats
def find_non_convertible_floats(series):

    non_convertibles = set()

    for item in series:
        try:
            float(item)

        except ValueError:
            non_convertibles.add(item)

    return non_convertibles


# Function to find all columns with int and float type values
def numeric_cols(dataframe):
    # Initialize a list to store column names with numeric data
    numeric_columns = []

    # Iterate through the columns of the DataFrame
    for column in dataframe.columns:
        # Check if the column contains numeric data (float or int)
        if pd.api.types.is_numeric_dtype(dataframe[column]):
            numeric_columns.append(column)
    
    return numeric_columns


# Function to find all null, duplicate and non convertible float values
def find_null(dataframe):
    
    # List of all numeric columns
    numeric = numeric_cols(dataframe)
    
    # Check if there are non assigned values in the dataframe
    null_count = dataframe.isna().sum()
    
    # Check for duplicates in the dataframe
    duplicate_count = dataframe.duplicated().sum()
    
    # Check for values that are not floats
    non_convertibles = [find_non_convertible_floats(dataframe[col]) for col in numeric]
    
    print('null count: ' + str(sum(null_count)))
    print('duplicate count: ' + str(duplicate_count))
    print('non convertible floats: ' + str(sum([len(x) for x in non_convertibles])))
    

# Function to drop all columns with 20% null values and drop all rows with null values
def drop_null(dataframe):
    
    # Calculate the percentage of null values in each column
    null_percentage = dataframe.isnull().mean() * 100
    
    # Filter columns where null percentage is greater than 20%
    columns_to_drop = null_percentage[null_percentage > 20].index
    
    # drop columns with more than 35% of null values
    dataframe = dataframe.drop(columns=columns_to_drop)
    
    # Drop rows with null values
    dataframe = dataframe.dropna()
    
    return dataframe
```

## Grouping: Geological parameters

<hr style="height: 4px; background-color: #2B547E;">
</h1>

To make accurate predictions based on the commodity, the well, and their geological characteristics, it is important to create group clusters. This has been done by clustering oil wells by geological characteristics Density Porosity (%) and Effective Porosity (%) as these characteristics are positively correlated with oil production, and negatively correlated to gas production. This counts as well for gas wells, where geological characteristics Clay Volume (%) and Water Saturation (%) as these characteristics are positively correlated with gas production and negatively correlated with oil production.

```python
# Filter out only producing wells
metaDF = metaDF.loc[metaDF['ENVWellStatus'] == 'PRODUCING']
```

## Oil

```python
# Create dataframe with oil wells
oilDF = metaDF.loc[metaDF['ENVWellType'] == 'OIL']
```

```python
# Selecting the relevant features for clustering
GeoOil = oilDF[['DensityPorosity_PCT','EffectivePorosity_PCT']]

# Scaling the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(GeoOil)

# Choosing the number of clusters (k) - this example uses k=3, but you may need to adjust this
k = 6
kmeans = KMeans(n_clusters=k, random_state=0)

# Fitting the K-Means model
kmeans.fit(X_scaled)

# Assigning the cluster labels to your DataFrame
oilDF['Cluster'] = kmeans.labels_

# Plot each cluster
for cluster_label in range(k):
    cluster_data = X_scaled[oilDF['Cluster'] == cluster_label]
    plt.scatter(cluster_data[:, 0], cluster_data[:, 1], label=f'Cluster {cluster_label + 1}')

# Plotting centroids
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=300, c='black', label='Centroids')

plt.title('Clustering of geological parameters')
plt.xlabel('Density Porocity Pct (Scaled)')
plt.ylabel('Effective Porocity Pct (Scaled)')
plt.legend()
plt.grid(True)
plt.show()
```

For the two characteristics used in the identification of wells, six clusters were made as five clusters would include clusters which have a high euclidean distance. With six clusters, there are more dense groups of wells based on their characteristics.

```python
# Selecting the relevant features for clustering
GeoGas = gasDF[['WaterSaturation_PCT','ClayVolume_PCT']]

# Scaling the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(GeoGas)

# Choosing the number of clusters (k) - this example uses k=3, but you may need to adjust this
k = 5
kmeans = KMeans(n_clusters=k, random_state=0)

# Fitting the K-Means model
kmeans.fit(X_scaled)

# Assigning the cluster labels to your DataFrame
gasDF['Cluster'] = kmeans.labels_

# Plot each cluster
for cluster_label in range(k):
    cluster_data = X_scaled[gasDF['Cluster'] == cluster_label]
    plt.scatter(cluster_data[:, 0], cluster_data[:, 1], label=f'Cluster {cluster_label + 1}')

# Plotting centroids
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=300, c='black', label='Centroids')

plt.title('Clustering of geological parameters')
plt.xlabel('Water Saturation Pct (Scaled)')
plt.ylabel('Clay Volume Pct (Scaled)')
plt.legend()
plt.grid(True)
plt.show()
```

## Grid Search

To find out what parameters must be used in the creation of the forecasting model per cluster, a function has been created which will iterate over different combinations of parameters being input into the model. This has been done over all clusters. The function returns a dataframe with all clusters and their best parameters based on the best accuracy (AIC).

```python
def find_best_sarima_params(dataframes):
    # Define the range of values for p, d, q (non-seasonal orders)
    p_values = range(0, 2)
    d_values = range(0, 2)
    q_values = range(0, 2)
    
    # Define the range of values for P, D, Q, m (seasonal orders)
    P_values = range(0, 2)
    D_values = [1]
    Q_values = range(0, 2)
    m_values = [12]  # Assuming monthly data with annual seasonality; adjust as needed
    
    results = []

    for df_name, df in dataframes.items():
        # Ensure the DataFrame has a datetime index with a defined frequency
        if df.index.inferred_freq is None:
            print(f"{df_name}: Index does not have a defined frequency. Attempting to infer...")
            df = df.asfreq(pd.infer_freq(df.index))
        
        df_train = df.sort_index().loc[:'2020']
        df_test = df.sort_index().loc['2021':]

        best_aic = float("inf")
        best_params = None

        # Create all combinations of SARIMA parameters
        param_combinations = list(itertools.product(p_values, d_values, q_values, P_values, D_values, Q_values, m_values))

        with warnings.catch_warnings():
            warnings.filterwarnings("ignore", category=ConvergenceWarning)
            
            for params in param_combinations:
                try:
                    model = sm.tsa.SARIMAX(df_train, 
                                           order=params[:3], 
                                           seasonal_order=params[3:],
                                           enforce_stationarity=False, 
                                           enforce_invertibility=False)
                    result = model.fit(disp=False)
                    aic = result.aic
                    
                    if aic < best_aic:
                        best_aic = aic
                        best_params = params

                except Exception as e:
                    continue
        
        results.append({
            'DataFrame': df_name,
            'Best Parameters': best_params,
            'Best AIC': best_aic
        })

    # Convert results to a DataFrame for better readability
    results_df = pd.DataFrame(results)
    return results_df
```

### Model on Oil

```python
oil_dataframes = {
    '0': pred_oilDF1,
    '1': pred_oilDF2,
    '2': pred_oilDF3,
    '3': pred_oilDF4,
    '4': pred_oilDF5,
    '5': pred_oilDF6
}

oil_results_df = find_best_sarima_params(oil_dataframes)
print(oil_results_df)
```
